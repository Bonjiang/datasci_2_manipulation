```python
import pandas as pd
import numpy as np
```

### Load data

```python
df = pd.read_csv('https://raw.githubusercontent.com/hantswilliams/HHA_507_2023/main/WK2/data/healthcare_data_cleaning.csv')
df
```

### Identify and Handling Missing Values in a Dataset

```python
df.replace('missing', np.nan, inplace=True)
```

```python
missing_values = df.isnull().sum()
missing_values
```

##### Checking for a specific column

```python
df['Visited Last Month'].isnull().sum()
```

##### Dropping rows with any missing values

```python
df.dropna(inplace=True)
df
```

 ##### *67406 rows were removed*

### Removing Duplicates

##### Removing duplicate rows

```python
df.drop_duplicates(inplace=True)
df
```

##### *1766 rows were removed*

##### Removing duplicate columns

```python
df = df.T.drop_duplicates().T
df
```

##### *0 columns were removed*

### Cleaning column names


##### Removing leading and trailing white space from a column

```python
df.columns = df.columns.str.strip()
df
```

##### Converting column names to uppercase

```python
df.columns = df.columns.str.upper()
df
```

### Creating New Columns Based on Existing Ones

##### Creating a 'Birth Date' column based on patient age and reference date

```python
reference_date = '2023-09-12'
df['BIRTH DATE'] = pd.to_datetime(reference_date) - pd.to_timedelta(df['PATIENT AGE'])
df
```

### Aggregating Data

```python
grouped_data = df.groupby('STATE OF RESIDENCE').agg({
    'PATIENT AGE': 'max'
})
grouped_data
```

```python
grouped_data = df.groupby('HAS INSURANCE').agg({
    'VISITED LAST MONTH': 'count'
})
grouped_data
```

### Computing Summary Statistics

```python
summary_statistics = df.describe()
summary_statistics
```

### Using Pivot Tables

```python
pivot_df = df.pivot_table(index='STATE OF RESIDENCE', columns='GENDER', values='NUMBER OF TESTS', aggfunc='count')
pivot_df
```

```python
pivot_df = df.pivot_table(index='HAS INSURANCE', columns='PREFERRED DOCTOR', values='VISITED LAST MONTH', aggfunc='count')
pivot_df
```
